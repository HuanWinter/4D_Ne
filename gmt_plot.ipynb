```python
import pygmt

fig = pygmt.Figure()
fig.basemap(region="g", projection="W10i", frame=True)
fig.coast(shorelines=True)
fig.show()
```

```python
fig = pygmt.Figure()
fig.basemap(region="g", projection="W10i", frame=True)
fig.coast(shorelines="0.5/0.5p,black")
fig.show()
```

```python
fig = pygmt.Figure()
fig.basemap(region="g", projection="W10i", frame=True)
fig.coast(land="#666666", water="skyblue")
fig.show()
```
