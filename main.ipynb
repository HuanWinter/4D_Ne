```python
## CPU multiprocessing

from multiprocessing import Pool 
from multiprocessing import cpu_count
#from functools import partial

cores = cpu_count()
Pool = pool(processes=cores)
cores

def read(str, ts, year_omni, month_omni, day_omni, hour_omni, \
     doy, Alt, Lat, Lon, Azi, TEC, Ne, X_omni, UT, i):
    start = time.clock()
    path = str[i]
    path = path.strip('\n')
    file = nc.Dataset(path)
    #print(file.variables.keys())
    #print(file.ncattrs)
    #for attr in file.ncattrs():
     #   print(attr)
    year = file.getncattr('year')
    month = file.getncattr('month')
    day = file.getncattr('day')
    UT[i] = file.getncattr('hour')
    length = len(file.variables['MSL_alt'][:])
    doy[i] = month + day/30-1
    #print(t)
    
    index_omni = [j for j in range(ts.data['time1'].size) if \
             year[i] == year_omni[j] \
             and month[i] == month_omni[j] \
             and day[i] == day_omni[j] \
             and UT[i] == hour_omni[j] ]
    X_omni[i][:] = np.array(ts.data['all1'][index_omni].sel(products=features).values)
    #X_t = np.hstack([X_omni[i], doy[i], UT[i]])
    #X_t = 

    #print(X_omni[i].size)
    
    #for j in range(length):
    Alt.append(file.variables['MSL_alt'][:])
    Lat.append(file.variables['GEO_lat'][:])
    Lon.append(file.variables['GEO_lon'][:])
    Azi.append(file.variables['OCC_azi'][:])
    TEC.append(file.variables['TEC_cal'][:])
    Ne.append(file.variables['ELEC_dens'][:])
    end = time.clock()
    print(i, 'th round, running time: ', end-start)
    
    return Alt Lat Lon Azi TEC Ne X_omni doy UT

pool = Pool()
func  = partial(read, str, ts, year_omni, month_omni, day_omni, hour_omni, \
     doy, Alt, Lat, Lon, Azi, TEC, Ne, X_omni, UT)
iter = i in range(len(str))
pool.map(func, iter)
pool.close()
pool.join()
```
