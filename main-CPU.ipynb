```python
# global value
def init():
    global data
    data = ts.data['time1']


    
def omni_read(ind):
    #global data 
    d_out = np.zeros(4,)
    time_index = data[ind]
    t = (time_index - np.datetime64('1970-01-01T00:00:00Z')) / np.timedelta64(1, 's')
    a = datetime.utcfromtimestamp(t)
    
    d_out[0] = np.int(a.year)
    d_out[1] = np.int(a.month)
    d_out[2] = np.int(a.day)
    d_out[3] = np.int(a.hour)
    
    return d_out
    #return a.year, a.month, a.day, a.hour

if __name__ == '__main__':
    
    #global data  
    
    data = ts.data['time1']
    start = time.time()
    pool = Pool(8,init,())
    #pool=Pool(8)
    
    index =  range(1,ts.data['time1'].size)    
    print(index)
        
    date = pool.map(omni_read, index)
    print(date[1000])
    
    end = time.time()
    print('running time: ', end-start)    
```
