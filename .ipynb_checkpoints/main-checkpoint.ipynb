```python
# read the RO file
path = '/export/scratch2/andong/Data/ROdata/cosmic2013/ionPrf/'

files = []
lists = open("cosmic_list.txt","w")
# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for file in f:
        if '.nc' in file:
            files.append(os.path.join(r, file))

for f in files:
    lists.write("%s\n"%f)
```
